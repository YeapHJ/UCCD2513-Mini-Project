3. Write codes that performs the following:
- Load the video “img_pexels.mp4” into the Python environment, resize it and display the videos with smaller frames (The frames can be of any size, as long as it is smaller). You can specify an arbitrary frame rate.
- Save it as a separate files: “smaller_img_pexels.avi” or "smaller_img_pexels.mp4"

```python
import cv2 as cv

video_path = "img_pexels.mp4"
cap = cv.VideoCapture(video_path)

fps = cap.get(cv.CAP_PROP_FPS)
width = int(cap.get(cv.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv.CAP_PROP_FRAME_HEIGHT))

new_width = int(width / 2)  
new_height = int(height / 2)  

output_path = "smaller_img_pexels.mp4"  
fourcc = cv.VideoWriter_fourcc(*"XVID")  
out = cv.VideoWriter(output_path, fourcc, fps, (new_width, new_height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    resized_frame = cv.resize(frame, (new_width, new_height))

    cv.imshow("Smaller Frames", resized_frame)
    if cv.waitKey(1) & 0xFF == ord('q'):
        break

    out.write(resized_frame)

cap.release()
out.release()
cv.destroyAllWindows()
```

4. Enlarge the image "dog.jfif" by using different techniques: 1) Linear interpolation, 2) Cubic interpolation and 3) nearest neighbor interpolation. Comment on the upscaled of all the outputs.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

image_path = "dog.jfif"
image = plt.imread(image_path)

enlargement_factor = 2  

linear_interpolated = ndimage.zoom(image, (enlargement_factor, enlargement_factor, 1), order=1)
cubic_interpolated = ndimage.zoom(image, (enlargement_factor, enlargement_factor, 1), order=3)
nearest_neighbor_interpolated = ndimage.zoom(image, (enlargement_factor, enlargement_factor, 1), order=0)

plt.subplot(131)
plt.title("Linear Interpolation")
plt.imshow(linear_interpolated)
plt.axis("off")

plt.subplot(132)
plt.title("Cubic Interpolation")
plt.imshow(cubic_interpolated)
plt.axis("off")

plt.subplot(133)
plt.title("Nearest Neighbor Interpolation")
plt.imshow(nearest_neighbor_interpolated)
plt.axis("off")

plt.tight_layout()
plt.show()
```
